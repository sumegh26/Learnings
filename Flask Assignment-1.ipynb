## Q1. What is Flask Framework? What are the advantages of Flask Framework?



Flask is a popular open-source web framework for building web applications in Python. It is a micro-framework, meaning it is designed to be simple and lightweight, with only the essential features needed to build a web application.

Some advantages of using Flask framework include:

Lightweight: Flask is a micro-framework, which means it is lightweight and easy to use. It provides a simple and minimalistic set of tools to build web applications.

Easy to Learn: Flask is easy to learn, especially for developers who are familiar with Python. Its intuitive API makes it easy to build web applications quickly and efficiently.

Flexibility: Flask is highly flexible, and developers can choose the components they need to build their web application. It does not impose a particular project structure, allowing developers to choose how they want to organize their application.

Modular: Flask is highly modular, and developers can use extensions to add additional functionality to their web application. This modular design also makes it easy to test and maintain Flask-based web applications.

Scalable: Flask is scalable, and it can be used to build small or large web applications. It can handle high traffic and heavy loads without compromising performance.

Large Community: Flask has a large and active community of developers who contribute to the framework and its ecosystem. This means that developers can easily find solutions to common problems and access a wide range of third-party extensions and libraries.

Overall, Flask is an excellent choice for developers who want a lightweight, flexible, and easy-to-learn web framework for building web applications in Python. Its simplicity and modularity make it a popular choice for small and large projects alike.

## Q3. What is App routing in Flask? Why do we use app routes?

App routing is a feature in the Flask web framework that allows developers to specify which code should be executed for which URL. In other words, app routing determines how Flask responds to different requests from the client.

In Flask, app routing is implemented using the @app.route decorator. This decorator is used to specify the URL for a particular function in the application.

We use app routes in Flask to handle different types of requests from the client. For example, we can define different routes for different pages in our web application or different API endpoints that return different types of data. App routing allows developers to create a logical structure for their web applications and handle different types of requests efficiently.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
## route to show the following details:
## Company Name: ABC Corporation
## Location: India
## Contact Detail: 999-999-9999
## Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask
app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return 'Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999'

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. This function generates a URL for a given function name and arguments.

Here's an example code that demonstrates the working of the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f'Hello {username}'

@app.route('/')
def index():
    user_url = url_for('profile', username='John')
    return f'User URL: {user_url}'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

In this example, the @app.route('/user/<username>') decorator specifies that the profile function should be executed when the user navigates to a URL in the form of /user/<username>. The profile function takes a username argument and returns a message that greets the user.

The @app.route('/') decorator specifies that the index function should be executed when the user navigates to the root URL of the application. The index function uses the url_for() function to generate a URL for the profile function with the argument username='John'. This generated URL is then returned to the browser as a message.

When the user navigates to http://localhost:5000/, the index function is executed, and the generated URL for the profile function is displayed as a message: "User URL: /user/John". This URL can be used to navigate to the profile function with the argument username='John'. When the user navigates to this URL (http://localhost:5000/user/John), the profile function is executed, and the message "Hello John" is displayed in the browser.
